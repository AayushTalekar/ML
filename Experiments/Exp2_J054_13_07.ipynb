21. Repeat step 19and20 for other combinations of independent variables wrt price. Observations? Inferences?

```
cols = df.describe(include=np.object).columns
redundant = []
for i in cols:
    redundant.append(i)
    for j in cols:
        if i==j or j in redundant:
            continue
        else:
            #print(i,' and ',' with respect to price')
            temp = df.groupby([i,j])['price'].mean()
            print(temp, end='\n\n')
            temp = df[[i,j,'price']]
            temp.columns=[i,j,'price']
            for_pivot = temp.groupby([i,j], as_index=False).mean()
            piv = for_pivot.pivot(index = i, columns = j)
            piv.replace(np.nan, '-', inplace= True)
            print('Pivot Table')
            print()
            print(piv)
            print()
            print('------', end='\n\n')
```

23. Calculate the pearson correlation between “wheel-base” and “price”. What can you conclude from p-value (Hint: use stats from scipy which has pearsonr())

```
from scipy import stats
```

```
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value) 
```

Since the p value for the correlation is less than 0.01, there is statistical correlation between wheel base and price

24. Perform one way ANOVA test using f_oneway() of stats to check if different groups of “drive-wheels” are correlated with “price”. What do you understand from F-test and p-value results?

```
anova_df=df[['drive-wheels', 'price']].groupby(['drive-wheels'])
```

```
f_val, p_val = stats.f_oneway(anova_df.get_group('fwd')['price'], anova_df.get_group('rwd')['price'], anova_df.get_group('4wd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

since p > 0.01, there is no correlation

```
f_val, p_val = stats.f_oneway(anova_df.get_group('fwd')['price'], anova_df.get_group('rwd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

since p < 0.01, there is correlation

```
f_val, p_val = stats.f_oneway(anova_df.get_group('fwd')['price'], anova_df.get_group('4wd')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

since p > 0.01 there is no correlation
