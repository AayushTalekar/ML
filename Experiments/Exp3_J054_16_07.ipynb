```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
```

```
sns.set_style('darkgrid')
```

```
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
 sheet_name='Canada by Citizenship',
 skiprows=range(20),
 skipfooter=2)
 
print ('Data read into a pandas dataframe!')
```

```
df_can.head()
```

```
df_can.info()
```

```
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```
df_can.head()
```

```
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Continent-Region', 'DevName':'Development Status'}, inplace=True)
```

```
df_can.head()
```

```
df_can['total migrants'] = df_can.sum(axis=1)
df_can
```

```
df_can.describe()
```

```
df_can.set_index('Country', inplace=True)
```

```
df_can.head()
```

```
years = list(map(int,range(1980,2014)))
india = df_can.loc['India',years]
```

```
india.plot()
```

```
#change the index to integer
india.index = india.index.map(int)

india.plot(kind='line')
plt.title('Immigration from India')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

#update the plot
plt.show()
```

```
years = list(map(int,range(1980,2014)))


chinaindia = df_can.loc[['China','India'], years]

new_chinaindia = chinaindia.transpose()
new_chinaindia.plot()
```

```
df_can.sort_values(by='total migrants', ascending=False, axis='index', inplace=True)

top5 = df_can.head(5)

top5
```

```
years = list(map(int,range(1980,2014)))


top5_clean = top5[years]

top5_clean = top5_clean.transpose()
top5_clean.plot(figsize=(18,6))
```

```
tail5 = df_can.tail(5)
years = list(map(int,range(1980,2014)))


tail5_clean = tail5[years]

tail5_clean = tail5_clean.transpose()
tail5_clean.plot(figsize=(18,6))
```

```
top5_clean.index = top5_clean.index.map(int)

top5_clean.plot(kind="area", stacked=False, figsize=(20,10), alpha=0.25)

plt.title('Immigration trends in top five countries (1980-2013)')
plt.xlabel('Years')
plt.ylabel('Immigrants count')

plt.show()
```

```
tail5_clean.index = tail5_clean.index.map(int)

tail5_clean.plot(kind="area", stacked=False, figsize=(20,10), alpha=0.25)

plt.title('Immigration trends in bottom five countries (1980-2013)')
plt.xlabel('Years')
plt.ylabel('Immigrants count')

plt.show()
```

```
top5_clean.index = top5_clean.index.map(int)

top5_clean.plot(kind="area", stacked=True, figsize=(20,10), alpha=0.25)

plt.title('Immigration trends in top five countries (1980-2013)')
plt.xlabel('Years')
plt.ylabel('Immigrants count')

plt.show()
```
